#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
import pandas as pd

df = pd.read_csv("datasets/users_dataset.csv", sep = "|")

# Seu código
df[["occupation", "age"]].groupby("occupation").mean()
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
# Seu código

#Funcao agg = aggregation
df.groupby(["occupation", "gender"]).agg({"age":["min", "max", "mean"]})
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
# Seu código
df.groupby(["occupation", "gender"]).agg({"age":["mean"]})
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
# Seu código
df.groupby(["occupation", "gender"]).count()["user_id"] / df.groupby(["occupation"]).count()["user_id"]
````
